```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
import numpy as np
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
model_lr=LogisticRegression()
model_SV=SVC()
model_RF=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
model_lr.fit(X_train,y_train)
```

```python
model_lr.score(X_test,y_test)
```

```python
model_SV.fit(X_train,y_train)
```

```python
model_SV.score(X_test,y_test)
```

```python
model_RF.fit(X_train,y_train)
```

```python
model_RF.score(X_test,y_test)
```

```python
def Model_score(model,X_train, X_test, y_train, y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold
```

```python
kf=KFold(n_splits=10)
```

```python
Model_score(LogisticRegression(),X_train,X_test,y_train,y_test)
```

```python
Model_score(SVC(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds=StratifiedKFold(n_splits=10)
```

```python
scores_lr=[]
scores_sv=[]
scores_rf=[]
for train_index,test_index in kf.split(digits.data):
    X_train,X_test,y_train,y_test = digits.data[train_index],digits.data[test_index],digits.target[train_index],digits.target[test_index]
    scores_sv.append(Model_score(SVC(),X_train,X_test,y_train,y_test))
    scores_lr.append(Model_score(LogisticRegression(max_iter=1000),X_train,X_test,y_train,y_test))
    scores_rf.append(Model_score(RandomForestClassifier(n_estimators=40),X_train,X_test,y_train,y_test))
```
